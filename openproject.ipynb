```python
from typing import Any
import requests
import json
import base64
from enum import Enum
```

```python
API_TOKEN = "82f6ea7c691739a265ea60c377bc5c2276a31871e475006b4b2a2f95f05e0463"
HOST = "http://openproject.lan"

PROJECT_IDENTIFIER = "requirements-gathering"
# POST /api/v3/projects/{project_identifier}/work_packages/form


```

```python
# Build the base64 Authorization header
auth_string = f"apikey:{API_TOKEN}"
auth_bytes = auth_string.encode('utf-8')
auth_b64 = base64.b64encode(auth_bytes).decode('utf-8')

headers = {
    "Authorization": f"Basic {auth_b64}",
    "Content-Type": "application/json",
    "Accept": "application/json",
}

session = requests.Session()
session.headers.update(headers)
```

```python
class Endpoint(Enum):
    LIST_WORK_PACKAGES = lambda: f"{HOST}/openproject/api/v3/work_packages/"
    GET_WORK_PACKAGE = lambda work_package_id: f"{HOST}/openproject/api/v3/work_packages/{work_package_id}"

    LIST_PROJECTS = lambda: f"{HOST}/openproject/api/v3/projects/"
    GET_PROJECT = lambda project_id: f"{HOST}/openproject/api/v3/project/{project_id}"
```

```python
response = session.get(Endpoint.LIST_PROJECTS())

project: dict[str, Any] | None = None
project_id: int | None = None
for project in response.json()["_embedded"]["elements"]:
    if project['identifier'] == PROJECT_IDENTIFIER:
        project = project
        project_id = project['id']
        break

print(f"Project ID: {project_id}")
print(json.dumps(project, indent=4))
```

```python
project_work_packages_endpoint = f"{HOST}{project['_links']['workPackages']['href']}"
response = session.get(project_work_packages_endpoint)

print(response.status_code)
print(json.dumps(response.json(), indent=4))
```
